```python
from bs4 import BeautifulSoup
import requests

url = 'https://www.pace.edu/find-your-program/graduate'
headers = {"User-Agent": "Mozilla/5.0"}
response = requests.get(url, headers=headers)

soup = BeautifulSoup(response.content, 'html.parser')

# Find all program buttons
program_buttons = soup.find_all('button', class_='accordion-trigger')

programs = []

for btn in program_buttons:
    # Get the program name from the span
    title_span = btn.find('span', class_='item-title')
    if title_span:
        program_name = title_span.get_text(strip=True)
        programs.append(program_name)

# Print the results
for program in programs:
    print(program)


# Find all anchor tags with the specific class
program_links = soup.find_all('a', class_='button--alt')

# Extract hrefs and print them
for link in program_links:
    href = link.get('href')
    if href:
        # Make full URL if needed
        full_url = f"https://www.pace.edu{href}"
        print(full_url)
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

# Target URL and headers
url = 'https://www.pace.edu/find-your-program/graduate'
headers = {"User-Agent": "Mozilla/5.0"}

# Send GET request
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, 'html.parser')

# Find all program titles
program_buttons = soup.find_all('button', class_='accordion-trigger')
program_titles = []
for btn in program_buttons:
    title_span = btn.find('span', class_='item-title')
    if title_span:
        program_titles.append(title_span.get_text(strip=True))

# Find all program links
program_links = []
anchors = soup.find_all('a', class_='button--alt')
for a in anchors:
    href = a.get('href')
    if href:
        full_url = f"https://www.pace.edu{href}"
        program_links.append(full_url)

# Combine data safely
length = min(len(program_titles), len(program_links))  # ensure alignment
data = {
    'Program Name': program_titles[:length],
    'Program Link': program_links[:length]
}

# Create DataFrame
df = pd.DataFrame(data)

# Save to CSV
df.to_csv('pace_graduate_programs.csv', index=False)

# Optional: Display first few rows
print(df.head())
```
